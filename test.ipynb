```python
import pandas as pd

url = "https://pl.wikipedia.org/wiki/Kamil_Grosicki"
tables = pd.read_html(url, header=0)

print(f"Liczba tabel: {len(tables)}")

for i, table in enumerate(tables):
    print(f"\nTabela {i}:")
    print(table.head())


nowa_tabela = tables[8].copy()
nowa_tabela.columns = nowa_tabela.iloc[0]
nowa_tabela = nowa_tabela[1:]
nowa_tabela
```

```python
# Zliczamy, ile razy 'Grał' jest puste lub NaN
full_90_count = (nowa_tabela['Grał'] == "90'").sum()
print(f"Ilość pełnych 90 minut: {full_90_count}")
```

```python
nowa_tabela.copy()['Data'] = pd.to_datetime(nowa_tabela['Data'], format='%d.%m.%Y', errors='coerce')

nowa_tabela['Data'] = pd.to_datetime(nowa_tabela['Data'], dayfirst=True)
last_3 = nowa_tabela.sort_values('Data', ascending=False).head(3)


last_3
```

```python
wynik_split = last_3['Wynik'].str.split(':', expand=True)
last_3['Bramki_strzelone'] = pd.to_numeric(wynik_split[0], errors='coerce')
last_3['Bramki_stracone'] = pd.to_numeric(wynik_split[1], errors='coerce')




import matplotlib.pyplot as plt
import numpy as np


width = 0.1
labels = last_3['Data'].dt.strftime('%d-%m-%Y').tolist()

x = np.arange(len(labels))
fig, ax = plt.subplots(figsize=(10,6))

rects1 = ax.bar(x - width, last_3['Bramki_strzelone'], width, label='Bramki strzelone', color='green')
rects2 = ax.bar(x, last_3['Bramki_stracone'], width, label='Bramki stracone', color='red')

ax.set_ylabel('Liczba')
ax.set_xlabel('Data')
ax.set_title('Wyniki i czas gry w 3 ostatnich meczach')
ax.set_xticks(x)
ax.set_xticklabels(labels)

ax.legend()

last_3
```

```python
def extract_first_number(s):
    # Przejdź znak po znaku i zbieraj cyfry aż do pierwszego nie-cyfrowego znaku
    num_str = ''
    for char in s:
        if char.isdigit():
            num_str += char
        elif num_str:  # jak już zacząłeś zbierać cyfry i napotkasz coś innego, przerywaj
            break
    return int(num_str) if num_str else 0


def minutes_played(x):
    if pd.isna(x):
        return 0
    x = str(x).strip()

    # Obsługa formatu '90+5' -> 95 minut
    if '+' in x:
        parts = x.replace("'", "").replace("do ", "").split('+')
        try:
            return int(parts[0]) + int(parts[1])
        except:
            return 0

    if x == "90'":
        return 90
    elif x.startswith('od '):
        start = extract_first_number(x)
        return 90 - start
    elif x.startswith('do '):
        end = extract_first_number(x)
        return end
    else:
        if x.endswith("'"):
            try:
                return int(x.replace("'", ""))
            except:
                return 0
        return 0



nowa_tabela['Minuty'] = nowa_tabela['Grał'].apply(minutes_played)
pd.set_option('display.max_rows', None)

display(nowa_tabela)
```
